This code is for downloading stack data from yahoo finance. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from yahoofinancials import YahooFinancials
```

```python
yf = YahooFinancials('AAPl')
start='2000-01-01'
end = '2021-01-01'
interval = 'weekly'
data = yf.get_historical_price_data(start_date=start,
                                   end_date=end,
                                   time_interval=interval)
aapl_df=pd.DataFrame(data['AAPL']['prices'])
aapl_df=aapl_df.drop('date',axis=1).set_index('formatted_date')
aapl_df.head()
```

Plot data.

```python
plt.style.use('seaborn')
aapl_df.plot(title="Apple's Stock Pprice")
```

Save data to a csv file

```python
aapl_df.to_csv('appleStock.csv',index=False)
```
